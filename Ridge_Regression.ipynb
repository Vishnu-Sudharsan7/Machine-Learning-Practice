```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('D:/vs/datset/Gold.csv')
data
```

```python
data=data.drop(['Date'],axis=1)
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split as tts
df_train,df_test=tts(data,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import Ridge
```

```python
y_train=df_train.pop('Close')
x_train=df_train
```

```python
model=Ridge(alpha=1.0)
```

```python
model.fit(x_train,y_train)
```

```python
accuracy=model.score(x_train,y_train)*100
accuracy #train data set accuracy
```

```python
y_test=df_test.pop('Close')
x_test=df_test
```

```python
prediction=model.predict(x_test)
```

```python
df=pd.DataFrame({'Actual':y_test,'Prediction':prediction})
df
```

```python
x_ax = range(len(x_test))
plt.scatter(x_ax, y_test, s=5, color="blue", label="original")
plt.plot(x_ax, prediction, lw=0.8, color="red", label="predicted")
plt.legend()
plt.show()
```

```python
acc=model.score(x_test,y_test)*100
acc
```
